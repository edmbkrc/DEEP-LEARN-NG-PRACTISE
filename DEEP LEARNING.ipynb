## 60 bin el yazması rakamların sınıflandırılması

```python
import tensorflow as tf
from tensorflow import keras
```

```python
mnist=keras.datasets.mnist
(train_images,train_labels), (test_images,test_labels)=mnist.load_data()
```

```python
train_images.shape
```

```python
train_images
```

```python
train_labels
```

```python
from matplotlib import pyplot as plt
plt.imshow(train_images[59999],cmap="gray");
```

```python
plt.imshow(train_images[590],cmap="gray");
```

```python
plt.imshow(train_images[23],cmap="gray");
```

```python
plt.imshow(train_images[897],cmap="gray");
```

```python
#Normalize
```

```python
train_images=train_images / 255.0
test_images=test_images / 255.0#normalize in farklı bir şekli
```

```python
#define model
#CNN- Convolutional Neural Network
model=keras.Sequential([
                       keras.layers.InputLayer(input_shape=(28,28)),
                       keras.layers.Reshape(target_shape=(28,28,1)),
                       keras.layers.Conv2D(filters=12,kernel_size=(3,3), activation=tf.nn.relu),
                       keras.layers.MaxPooling2D(pool_size=(2,2)),
                       keras.layers.Flatten(),
                       keras.layers.Dense(50)
    
])
```

```python
model.compile(optimizer="adam",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

```python
model.fit(train_images, train_labels, validation_split=0.1, epochs=10)
```

```python
#CNN özellikle el yazması yazılar için üretildiği  için 150 kere gitmesine 
#gerek kalmadan 10 gitgelle  öğrenebiliyor
```

```python
_, accuracy=model.evaluate(test_images, test_labels)
print("Accuracy: %.2f" % (accuracy*100))
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fashion_mnist=tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels)= fashion_mnist.load_data()
```

```python
train_images.shape
```

```python
 class_names= ["T-shirt/top", "Trousers", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
plt.figure()
plt.imshow(train_images[0], cmap="gray")
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
model=tf.keras.Sequential([
                           tf.keras.layers.Flatten(input_shape=(28,28)),
                           tf.keras.layers.Dense(128, activation="relu"),
                           tf.keras.layers.Dense(128, activation="relu"),
                           tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer="adam",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

```python
model.fit(train_images, train_labels, epochs=10, verbose=1)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print("\nTest accuracy: ", test_acc)
```

```python
model_json=model.to_json()
with open("model_minst.json", "w") as json_file:
    json_file.write(model_json)

model.save_weights("model_mnist.h5")
print("Saved model to disk")
```
